```python
import numpy as np
import matplotlib.pyplot as plt
ReLu=lambda x:np.max(x,0)#lambda x:x*(x>0)
LReLu= lambda x:x*(x>0)+0.01*x*(x<0)
sigmoid= lambda x: 1/(1+np.exp(-x))
tanh = lambda x:2*sigmoid(x)
softplus = lambda x:np.log(np.exp(x)+1)
```

```python
a=np.array([-1,1,-1,4,0,100,-5])
```

```python
print(softplus(a))
```
